```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat_cols=['mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']
for cols in cat_cols:
  data[cols]=le.fit_transform(data[cols])
```

```python
x=data[['area','bedrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=30,test_size=0.3)
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)
```

```python
print("MSE: ",mean_squared_error(y_test,y_pred))
print("MAE: ",mean_absolute_error(y_test,y_pred))
print("r2 Score: ",r2_score(y_test,y_pred))
```

#DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=4, random_state=41)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("MSE: ",mean_squared_error(y_test,y_pred))
print("MAE: ",mean_absolute_error(y_test,y_pred))
print("r2 Score: ",r2_score(y_test,y_pred))
```

#RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,max_depth=6,min_samples_split=10)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("MSE: ",mean_squared_error(y_test,y_pred))
print("MAE: ",mean_absolute_error(y_test,y_pred))
print("r2 Score: ",r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],         # 3
    'max_depth': [4, 6, 8, 10],              # 4
    'min_samples_split': [2, 5, 10],         # 3
    'min_samples_leaf': [1, 2],              # 2
    'max_features': ['sqrt', 'log2']         # 2
}
# Total combinations: 3 × 4 × 3 × 2 × 2 = 144
# With cv=4 → 144 × 4 = 576 fits

grid_search=GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='r2',
    cv=4,
    verbose=1,
    n_jobs=-1
)
grid_search.fit(x_train,y_train)
best_model=grid_search.best_estimator_
y_pred=best_model.predict(x_test)
print("Best Parameters:",grid_search.best_params_)  #showcases best parameters
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

#XGBoost

```python
from xgboost import XGBRegressor
model=XGBRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(f"MSE : {mse}")
print(f"MAE : {mae}")
print(f"r2_socre: {r2}")
```

```python
param_grid={
    'n_estimators':[100,150,200,300,500],
    'max_depth':[3,4,5,6,8,10],
    'learning_rate':[0.01,0.02,0.05,0.1,0.2],
    'subsample':[0.6,0.8,1.0]
}
grid_search=GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='r2',
    cv=4,
    verbose=1,
    n_jobs=-1
)
grid_search.fit(x_train,y_train)
best_model=grid_search.best_estimator_
y_pred=best_model.predict(x_test)
print("Best Parameters:",grid_search.best_params_)  #showcases best parameters
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
